## 1. Import Library

```python
import pandas as pd 
from matplotlib import pyplot as plt 
from sklearn.tree import DecisionTreeClassifier , plot_tree
```

## 2. Membaca Dataset

```python
df = pd.read_csv ('data_spesifikasi_mobile.csv')
df
```

## 3. Preprocessing Data

### 3.1 Menghapus Kolom yang Tidak Relevan

```python
df = df.drop('User ID', axis =1)
df
```

### 3.2 Pengecekan dan Penanganan Missing Values

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.isnull().sum()
```

### 3.3 Konversi Data Kategorikal ke Numerik

```python
df.replace({'Male': 0, 'Female':1}, inplace=True)
df
```

```python
df.replace({'Android': 0, 'iOS':1}, inplace=True)
df
```

```python
df['Device Model'].unique()
```

```python
df.replace({'Google Pixel 5': 0, 'OnePlus 9':1, 'Xiaomi Mi 11':2, 'iPhone 12':3,'Samsung Galaxy S21':4, 'OnePlus 9+B5':5}, inplace=True)
df
```

### 3.4 Mengubah Tipe Data untuk Konsistensi

```python
df['Screen On Time (hours/day)'] = df['Screen On Time (hours/day)'].astype(int)
df
```

## 4. Memisahkan Dataset Menjadi Fitur (X) dan Target (y) 

```python
x = df.drop('User Behavior Class', axis=1)
y = df['User Behavior Class']
```

## 5. Membagi Dataset untuk Pelatihan dan Pengujian 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y,test_size = 0.4)
```

## 6. Membuat dan Melatih Model 

```python
model= DecisionTreeClassifier (max_depth = 4)
model.fit (x_train, y_train)
```

## 7. Evaluasi Model

```python
y_pred= model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)*100
r2
```

## 8. Visualisasi Decision Tree

```python
class_labels = {
    1: "Light usage",
    2: "Medium-low usage",
    3: "Medium usage",
    4: "High usage",
    5: "Very high usage"
}
```

```python
def replace_class_labels(value):
    return [class_labels.get(i, str(i)) for i in value]
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(12, 7))
plot_tree(
    model, 
    filled=True, 
    class_names=['1', '2', '3', '4', '5'],
    feature_names=[
        'User ID', 
        'Device Model', 
        'Operating System', 
        'App Usage Time (min/day)', 
        'Screen On Time (hours/day)', 
        'Battery Drain (mAh/day)', 
        'Number of Apps Installed', 
        'Data Usage (MB/day)', 
        'Age', 
        'Gender'
    ] 
)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plot_tree(
    model, 
    filled=True, 
    class_names=replace_class_labels([1, 2, 3, 4, 5]),
    feature_names=[
        'User ID', 
        'Device Model', 
        'Operating System', 
        'App Usage Time (min/day)', 
        'Screen On Time (hours/day)', 
        'Battery Drain (mAh/day)', 
        'Number of Apps Installed', 
        'Data Usage (MB/day)', 
        'Age', 
        'Gender'
    ],
    rounded=True
)
plt.show()
```
